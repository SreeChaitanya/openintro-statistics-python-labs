```python
import pandas as pd 
df = pd.read_csv('atheism.csv')
df.head()
```

```python
response_tab = pd.crosstab(index=df["response"],  # Make a crosstab
                              columns="count")      # Name the count column
response_tab
```

```python
response_tab/response_tab.sum() #Proportion of the data that belongs to each category.
```

```python
response_nationality = pd.crosstab(index=df['response'],columns=df['nationality'])
response_nationality
```

```python
response_nationality = pd.crosstab(index=df['response'],columns=df['nationality'],margins=True)
response_nationality
```

```python
#chi2 test for independence
from scipy.stats import chi2_contingency
chi2_contingency(response_nationality)
#low value of p suggests significant dependence
```

```python
import numpy as np
np.random.seed(10)

# Sample data randomly at fixed probabilities
voter_race = np.random.choice(a= ["asian","black","hispanic","other","white"],
                              p = [0.05, 0.15 ,0.25, 0.05, 0.5],
                              size=1000)

# Sample data randomly at fixed probabilities
voter_party = np.random.choice(a= ["democrat","independent","republican"],
                              p = [0.4, 0.2, 0.4],
                              size=1000)

voters = pd.DataFrame({"race":voter_race, 
                       "party":voter_party})

voter_tab = pd.crosstab(voters.race, voters.party, margins = True)

voter_tab.columns = ["democrat","independent","republican","row_totals"]

voter_tab.index = ["asian","black","hispanic","other","white","col_totals"]

observed = voter_tab.iloc[0:5,0:3]   # Get table without totals for later use
voter_tab
```
