```python
import pandas as pd
df = pd.read_csv('bdims.csv')
df.head()
```

```python
#Splitting the dataframe based on gender
df_male = df[df['sex']==1]
df_female = df[df['sex']==0]
```

```python
fm_height_mean = df_female['hgt'].mean()
fm_height_sd =  df_female['hgt'].std()
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.distplot(df_female['hgt'])
plt.show()


```

```python
from scipy.stats import probplot
probplot(df_female['hgt'],plot=plt)
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(df_female['hgt'])
plt.show()
```

```python
y=norm.pdf(df_female['hgt'].sort_values(),165,fm_height_sd)

plt.plot(df_female['hgt'].sort_values(),y,'-o')
plt.hist(df_female['hgt'],normed=1)
plt.show()
```

```python
female = plt.scatter(df_female['hgt'],df_female['wgt'],color='blue')
male = plt.scatter(df_male['hgt'],df_male['wgt'],marker='x',color='red')
plt.legend((female,male),('Female',('Male')))
plt.title('ScatterPlot')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
pp_fm=sm.ProbPlot(df_female['hgt'].sample(200))
pp_m = sm.ProbPlot(df_male['hgt'].sample(200))
sm.qqplot_2samples(pp_fm,pp_m)
plt.show()
```
