```python
#Confidence intervals
sampl1 = area.sample(100)
sampl1.std()
```

```python
se = sampl1.std()/100
se
```

```python
lower = sampl1.mean()-(1.96*se)
upper = sampl1.mean()+(1.96*se)
#confidence interval (lower,upper)
(lower,upper)
```
